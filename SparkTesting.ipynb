```python
%python
# Cell 1: Sample store sales data
data = [
    ("Laptop", "Electronics", 1200, 2),
    ("Phone", "Electronics", 800, 5),
    ("Hoodie", "Apparel", 60, 10),
    ("Jeans", "Apparel", 80, 4),
    ("Sneakers", "Apparel", 120, 3),
    ("Blender", "Kitchen", 150, 1)
]

columns = ["Item", "Category", "Price", "Quantity"]

# Creating the DataFrame
df = spark.createDataFrame(data, schema=columns)
df.printSchema() #
display(df)
```

```python
%python
df.write.mode("overwrite").saveAsTable("sample_sales_data")
print("Table 'sample_sales_data' created")
```

```python
%python
# In a Python cell:
df_tables = spark.table("workspace.default.sample_sales_data")
display(df_tables)
```

```python
# Cell 2: Chaining transformations
from pyspark.sql.functions import col
# Spark only updates its logical execution plan here; no heavy lifting is done yet!
transformed_df_tables = df_tables \
    .filter(col("Price") > 50) \
    .withColumn("Total_Revenue", col("Price") * col("Quantity")) \
    .groupBy("Category") \
    .sum("Total_Revenue")

print("Transformations staged successfully!")
```

```python
# Cell 3: Running Actions

# Action 1: count() - Tells us how many rows are in our final transformation result
row_count = transformed_df_tables.count()
print(f"Total processed categories: {row_count}\n")

# Action 2: display() - A Databricks-exclusive action that renders a clean interactive table
display(transformed_df_tables)
```
